Q2. What is the difference between the mean, median, and mode? How are they used to measure the
central tendency of a dataset?

Ans: 
* Mean is the average value of the given observations.

Mean (xÌ…)  = Sum of Values / Number of Values

Mean of the data 10, 30, 40, 20, 50 is

Mean = (10 + 30 + 40 + 20+ 50) / 5

         = 30
         
         
         

* Median is the middle value of the given observations. 

If the number of values (n value) in the data set is odd then the formula to calculate the median is,

Median = [(n + 1)/2]th term

If the number of values (n value) in the data set is even then the formula to calculate the median is:

Median  = [(n/2)th term + {(n/2) + 1}th term] / 2

Median of the data 30, 40, 10, 20, 50 is,

10, 20, 30, 40, 50

Median = [(n + 1)/2]th term

Median = [(5 + 1)/2]th term

            = 30

* Mode is the most repeated value in the given observation.

Given set is {1, 2, 2, 2, 3, 3, 4, 5}

Mode = 2

Q5. How are measures of dispersion such as range, variance, and standard deviation used to describe
the spread of a dataset? Provide an example.

Ans: 
1) The range of a dataset is the difference between the largest and smallest values in that dataset. For example, in the two datasets below, dataset 1 has a range of 20 â€“ 38 = 18 while dataset 2 has a range of 11 â€“ 52 = 41. Dataset 2 has a broader range and, hence, more variability than dataset 1.

2) Variance is the average squared difference of the values from the mean. Unlike the previous measures of variability, the variance includes all values in the calculation by comparing each value to the mean.
While higher values of the variance indicate greater variability, there is no intuitive interpretation for specific values.



3) The standard deviation is the standard or typical difference between each data point and the mean. When the values in a dataset are grouped closer together, you have a smaller standard deviation. On the other hand, when the values are spread out more, the standard deviation is larger because the standard distance is greater.

Q8. What do you understand about skewness in data?

Ans: Skewness is a measure of the asymmetry of a distribution. A distribution is asymmetrical when its left and right side are not mirror images. A distribution can have right (or positive), left (or negative), or zero skewness.

Q9. If a data is right skewed then what will be the position of median with respect to mean?

Ans: If the distribution of data is skewed to the right, the mode is often less than the median, which is less than the mean.

Q12. For a normal distribution data what is the relationship between its measure of central tendency?

Ans: Any normal distribution has a graph that is perfectly symmetric about a vertical line through its peak. 

Therefore, all measures of central tendency (most commonly, the mean, median, and mode) give the same answer: the x -value of the peak.

numbers, so one must find the mean of these two values, 85 and 90.
 median = 87.5

* Mode: The score 90 occurs more frequently than the other
